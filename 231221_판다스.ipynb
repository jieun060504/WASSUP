다중 인덱스

행이나 열에 여러 계층을 가지는 인덱스 즉, 다중 인덱스(multi-index)를 설정할 수도 있다.
데이터프레임을 생성할 때 columns 인수에 다음 예제처럼 리스트의 리스트(행렬) 형태로 인덱스를 넣으면 다중 열 인덱스를 가지게 된다.

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2), #randn 정규분포를 이루는 실수
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])#칼럼 2차원 리스트
df3
#A, B 생략된것
```

```python
df3.columns.names = ["Cidx1", "Cidx2"] # 1, 2레벨 칼럼의 이름들
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"], #1레벨 인덱스
                          ["id_" + str(i + 1) for i in range(3)] * 2]) #for문 컴프리헨션, 2레벨 인덱스
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"], #1레벨 인덱스
                          ["id_" + str(i + 1) for i in range(3)] * 2]) #for문 컴프리헨션, 2레벨 인덱스
df4.columns.names = ["A/B", "C/D"]
df4.index.names = ["M/F", "ID"]
df4
```

```python
#for문으로 풀기
df_score3 = pd.DataFrame(columns=['반', '번호', '국어', '영어', '수학'])
df_score3
```

```python
# 데이터 입력
for i in range(5):
    df_score3.loc[i] = ['A', i+1] + list(np.random.randint(0, 101, size=3))
    df_score3.loc[i+5] = ['B', i+1] + list(np.random.randint(0, 101, size=3))
df_score3
```

```python
# 데이터 입력
for i in range(5):
    df_score3.loc[i] = ['A', i+1] + list(np.random.randint(0, 101, size=3))
    df_score3.loc[i+5] = ['B', i+1] + list(np.random.randint(0, 101, size=3))
df_score3 #AB 순환
```

```python
# 2번 인덱스 설정
df_score4 = df_score3.set_index(['반', '번호'])
df_score4
```

```python
#3번 각 학생의 평균 계산
df_score4['평균'] = round(df_score4.mean(axis=1),1)
df_score4
```
